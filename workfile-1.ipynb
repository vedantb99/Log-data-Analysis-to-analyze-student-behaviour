# Optimizing the Sequences
As we can observe the sequences obtained from above are very long and have too many repitions which will lead to problems while passing through HMM so we reduce the sequences and add other essential factors.

```python
from itertools import groupby 
short_sequences={}


# for student in student_list:
#     short_sequences[student]=[]
# for student in student_sequences:
#     test_list=student_sequences[student]
#     res = [i[0] for i in groupby(test_list)]
#     short_sequences[student]=res
    
for student in student_list:
    short_sequences[student]=[]
for student in student_sequences:
    new_res=[]
    test_list=student_sequences[student]
    for k, g in groupby(test_list):
#         print(k, sum(1 for _ in g))
        l=sum(1 for _ in g)
        if(l>4):
            new_res.append(k)
            new_res.append(k)
        elif(l<5):
            new_res.append(k)
        

    short_sequences[student]=new_res
        
new_sequences={}


for student in student_list:
    new_sequences[student]=[]
for student in new_sequences:
    test_list=student_sequences[student]
    new_res=[]
    res = [i[0] for i in groupby(test_list)]
    for k, g in groupby(test_list):
#         print(k, sum(1 for _ in g))
        l=sum(1 for _ in g)
        if(l>4):
            new_res.append(k)
            new_res.append(k)
        elif(l<5):
            new_res.append(k)
        

    new_sequences[student]=new_res
#Running Redo or RA operations on Short Sequences
redo_sequences={}
for student in student_list:
    redo_sequences[student]=[]
for student in new_sequences:
    i=0
    while i < (len(short_sequences[student])-2):
#         print(i,short_sequences[student][i])
        if((short_sequences[student][i+1]=='SV') and (short_sequences[student][i+2]==short_sequences[student][i])):
            redo_sequences[student].append('REDO')
            i=i+3
   
        elif((short_sequences[student][i]=='RA') and (short_sequences[student][i+2]=='SV')):
            redo_sequences[student].append('RADO')
            i=i+3
        elif(i!=0 and (short_sequences[student][i]=='RA') and (short_sequences[student][i+1]==short_sequences[student][i-1])):
            redo_sequences[student].append('RADO')
            i=i+2      
        else:
            redo_sequences[student].append(short_sequences[student][i])
            i=i+1
                
#Finalizing shortest sequence for HMM states

for student in redo_sequences:
    x="SV"
    if(x in redo_sequences[student]):
        res = list(filter((x).__ne__, redo_sequences[student]))
        redo_sequences[student]=res
#         print(redo_sequences['117A1004'])
length=0
for student in redo_sequences:
    length=max(length,len(redo_sequences[student]))
    if(length==len(redo_sequences[student])):
        stud=student
# print("The longest sequence is of length: ",length,stud)

problem_list=[]
for student in redo_sequences:
#     print(student,len(redo_sequences[student]))
    if(len(redo_sequences[student])<=8):
        problem_list.append(student)
print(problem_list)
```
